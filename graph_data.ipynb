```python
import osmnx as ox
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

seattle_graph = ox.graph_from_place(
    "Seattle, Washington, USA",
    network_type="walk"
)
seattle_graph = ox.project_graph(seattle_graph)

ox.save_graphml(seattle_graph, "seattle.graphml")

print("Graph downloaded!")
```

```python
nodes, edges =  ox.graph_to_gdfs(seattle_graph)
```

```python
df= pd.read_csv('seattle_dataset.csv')
```

```python
df['geometry'] = df.apply(lambda row: Point(row['longitude'],row['latitude']), axis=1)
df.head()
```

```python
barriers = gpd.GeoDataFrame(df, geometry="geometry",crs = "EPSG:4326")
barriers
```

```python
barriers = barriers.to_crs(edges.crs)
barriers
```

```python
joined = gpd.sjoin_nearest(
    barriers,
    edges,
    how="left",
    distance_col="dist",
    max_distance=20
)
```

```python
joined.columns
```

```python
joined = joined.sort_values("dist").drop_duplicates(subset="Unnamed: 0", keep='first')
joined
```

```python
edge_features = joined.groupby(["u","v","key"]).agg({
    "severity": ["mean","max","count"],
    "is_temporary": "mean",
    "length": "first"
})
```

```python
edge_features.rename(columns={'mean':'avg_severity', 'max': 'max_severity','count':'num_barriers'}, level=1,inplace=True)
edge_features.drop('is_temporary', axis=1, inplace=True)
edge_features.reset_index(inplace=True)
```

```python
edge_features.columns = [
    "".join(col).strip("") if isinstance(col, tuple) else col
    for col in edge_features.columns
]
print(edge_features.columns)
```

```python
label_counts = (
    joined
    .groupby(["u","v","key","label_type"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)
```

```python
dummy = edge_features.merge(
    label_counts,
    on=["u","v","key"],
    how="left"
)
dummy
```

```python
joined
```

```python
label_stats = (
    joined
    .groupby(["u","v","key","label_type"])
    .agg(
        count=("severity","count"),
        avg=("severity","mean"),
        max=("severity","max")
    )
    .reset_index()
)
```

```python
label_stats
```

```python
label_stats_wide = (
    label_stats
    .set_index(["u","v","key","label_type"])
    .unstack(fill_value=0)
)
```

```python
label_stats_wide
```

```python
label_stats_wide.columns = [
    f"{label}_{metric}"
    for metric, label in label_stats_wide.columns
]

label_stats_wide = label_stats_wide.reset_index()
label_stats_wide
```

```python
sai = edge_features.merge(
    label_stats_wide,
    on=["u","v","key"],
    how="left"
)

sai = sai.fillna(0)
sai
```

```python
sai.loc[47603]
```

```python
nodes_ll = nodes.to_crs(epsg=4326)

sai["u_lat"] = sai["u"].apply(lambda x: nodes_ll.loc[x].geometry.y)
sai["u_lon"] = sai["u"].apply(lambda x: nodes_ll.loc[x].geometry.x)

sai["v_lat"] = sai["v"].apply(lambda x: nodes_ll.loc[x].geometry.y)
sai["v_lon"] = sai["v"].apply(lambda x: nodes_ll.loc[x].geometry.x)
```

```python
sai.drop(['u','v'],axis=1, inplace=True)
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# your dataframe

# load neighborhoods
hoods = gpd.read_file("seattle_neighborhoods.geojson").to_crs("EPSG:4326")

# convert df → GeoDataFrame
gdf = gpd.GeoDataFrame(
    sai,
    geometry=gpd.points_from_xy(sai["u_lon"], sai["u_lat"]),
    crs="EPSG:4326"
)

# spatial join
out = gpd.sjoin(gdf, hoods, how="left", predicate="within")

# keep only what you want
out["neighborhood"] = out["name"]   # or NAME / neighborhood depending on file
out = out.drop(columns=["geometry", "index_right"])

print(out)
```

```python
sai = out[['severityavg_severity', 'severitymax_severity',
       'severitynum_barriers', 'lengthfirst', 'CurbRamp_count',
       'NoCurbRamp_count', 'NoSidewalk_count', 'Obstacle_count',
       'Occlusion_count', 'Other_count', 'SurfaceProblem_count',
       'CurbRamp_avg', 'NoCurbRamp_avg', 'NoSidewalk_avg', 'Obstacle_avg',
       'Occlusion_avg', 'Other_avg', 'SurfaceProblem_avg', 'CurbRamp_max',
       'NoCurbRamp_max', 'NoSidewalk_max', 'Obstacle_max', 'Occlusion_max',
       'Other_max', 'SurfaceProblem_max', 'u_lat', 'u_lon', 'v_lat', 'v_lon',
       'nhood']]
sai.to_csv("gnn_input.csv")
```
