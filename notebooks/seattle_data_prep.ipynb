```python
import json
import pandas as pd
with open('attributes-seattle.json', 'r') as file:
    data = json.load(file)
    df = pd.json_normalize(data)
    df = pd.json_normalize(df['features'][0])
```

```python
df.head()
```

```python
df['longitude'] = list(map(lambda x: x[0], df['geometry.coordinates']))
df['latitude'] = list(map(lambda x: x[1], df['geometry.coordinates']))
```

```python
df.drop(['geometry.coordinates'],axis=1, inplace=True)
```

```python
df.columns = ['type','geometry_type','attribute_id','label_type','neighborhood','severity','is_temporary','longitude','latitude']
df.head()
```

```python
df.to_csv('seattle_dataset.csv')
```
